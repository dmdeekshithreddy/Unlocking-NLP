```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
documents = [
    "Short text",
    "Somewhat longer text with more words",
    "This is a much longer document. It contains many words, indeed a lot more words than the others."
]
```

```python
vectorizer = CountVectorizer()
```

```python
# fit the vectorizer to the training documents
vectorizer.fit(documents)
```

```python
# print the vocabulary
# output is a dict where key is the 'word itself' and value is its 'index'
available_vocabulary = vectorizer.vocabulary_
print(type(available_vocabulary))
print(available_vocabulary)
```

```python
# print all the distinct words/ features from the vocabulary
distinct_words = vectorizer.get_feature_names_out()
print(len(distinct_words))
print(distinct_words)
```

```python
# trasform training documents
Xtrain = vectorizer.transform(documents)
print(Xtrain)

# each array from the dense matrix below is a document 
# so first record in sparse matrix "(0, 11)   1" is nothing but
# document wiht index 0 (outer array index), element with index 11  has a value of 1
```

```python
# Convert to dense matrix, to see the actual values including zeroes
Xtrain_dense_matrix = Xtrain.toarray()
print(type(Xtrain_dense_matrix))
print(Xtrain_dense_matrix)
```

```python
# Visual Representation
import matplotlib.pyplot as plt

plt.spy(Xtrain, markersize=5)
plt.title('Sparse Matrix Pattern')
plt.show()
```

#### Note:

* When working with very large sparse matrices, converting to a dense format or attempting to print all elements can be very inefficient and consume a significant amount of memory.

* Always consider the size and sparsity of your matrix when choosing how to view it, to avoid performance issues.
